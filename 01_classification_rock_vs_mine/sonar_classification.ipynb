# Rock vs Mine binary classification



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('sonar_data.csv', header=None)
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data[60].value_counts()
```

M = Mine

R = Rock

```python
# Separate features and labels
X = data.drop(columns=60, axis=1)
y = data[60]
```

```python
print(X.shape)
print(y.shape)
print(X.head())
print(y.head())
```

## Split into train and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

```python
print(y_train)
print(y_test)
```

## Model training

```python
model = LogisticRegression()
```

```python
# train the model 
model.fit(X_train, y_train)
```

### Question 2:
#### What happens to your model if you fit (train) it more than once on your training data ?

## Model evaluation

```python
y_prediction = model.predict(X_test)
```

```python
test_data_accuracy = accuracy_score(y_prediction, y_test)
print('test_data_accuracy: ', test_data_accuracy)
```

```python
print(sum(y_test==y_prediction)/len(y_test))
```

```python
print("accuracy in test data: ", test_data_accuracy)
```
